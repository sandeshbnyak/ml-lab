```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = "5AI.csv" 
df = pd.read_csv(filepath)
print(df)
```

```python
c1 = "Internal"
c2 = "External" 
selecteddata = df[[c1, c2]]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(selecteddata[c1], selecteddata[c2], color="blue", alpha=0.7)
plt.title(f"Scatter Plot of {c1} vs {c2}")
plt.xlabel(c1)
plt.ylabel(c2)
plt.grid(True)
plt.show()
```

```python
correlationcoefficient = np.corrcoef(selecteddata[c1],selecteddata[c2])[0, 1]
print(f"Pearson Correlation Coefficient ({c1} vs {c2}):{correlationcoefficient}")
```

```python
covmatrix = np.cov(selecteddata[c1], selecteddata[c2])
print("\nCovariance Matrix:")
print(covmatrix)
```

```python
correlationmatrix = selecteddata.corr()
print("\nCorrelation Matrix:")
print(correlationmatrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlationmatrix, annot=True, cmap="coolwarm", fmt=".2f",
cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
